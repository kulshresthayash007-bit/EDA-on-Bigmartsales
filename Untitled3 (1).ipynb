##### Step 4. Perform Univanate Analysis with distribution plots/countplots Univariate analysis is the simplest form of data analysis. It involves analysing data with only one variable to understand its clistribution, central tendency, and dispersion

```python
### Distribution plots for numerical columns
for num in numerical_columns:
    if num !="Outlet_Establishment_Year":
        sns.displot(data = df, x = num, bins = 15, height = 3, aspect = 2, kde = True)
        plt.axvline(df[num].mean(), label = "mean")
        plt.axvline
        plt.title(f'Distribution plot of {num}')
        plt.show()
```

- Item outlet sales and item visibility is positively skewed
- Not a clear pattern is visible in item_weight although the data seems symetrical 
- Majority of the weights around 12 units
- MRP values seems to follow approx. uniform distribution

```python
## Create countplot for all categorical columns including establishment year
```

```python
for cat in categorical_columns:
    if cat != "Item_Identifier":
        plt.figure(figsize = (5,5))    
        sns.countplot(data = df, x = cat, hue = cat)
        plt.title(f"Count plots of {cat}")
        plt.xticks( rotation = 90)
        plt.show()
```
